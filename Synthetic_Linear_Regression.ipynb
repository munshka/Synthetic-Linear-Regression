```python
import random

array = []
for _ in range(1000):

    x1 = random.uniform(-1000.0, 1000.0)
    x2 = random.uniform(-1000.0, 1000.0)
    x3 = random.uniform(-1000.0, 1000.0)
    y =  500*x1 - 413*x2 + 17*x3 +203
    if y >= 0:
      y=1
    else:
      y=-1

    array.append([x1, x2, x3, y])

print(array)
print(len(array))
```
